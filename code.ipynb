## Sales Data Analysis

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

# Data Cleaning

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1, inplace = True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

# Exploratory Data Anlaysis

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,4)})
sns.barplot(x ='Gender', y= 'Amount',data = sales_gen)
```

### From above graph we can observe that most buyers are females.

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

### From above graph we can observe that most of the buyers are of age group between 26-35 yrs female

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state,x='State',y='Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state,x='State',y='Amount')
```

### From above graph we can observe that most of the orders are from Uttar pradesh, Maharshtra and Karnataka respectively in terms of amount also.

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_val = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_val,x= 'Marital_Status',y='Amount',hue = 'Gender')
```

### From above two graphs we can observe that most purchase was done by married(female).

```python
ax = sns.countplot(data = df, x= 'Occupation')
sns.set(rc={'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_sec = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_sec,x='Occupation',y= 'Amount')
```

### From above graph we can observe that most buyers are from IT sector, Healtcare and Aviation.

```python
ax = sns.countplot(x= 'Product_Category', data = df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat = df.groupby(['Product_Category'], as_index=False).sum().sort_values(by= 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_cat,x= 'Product_Category', y= 'Amount')
```

### From above graph we can observe that most sold categories sold are food, clothing and electronics.

```python
sales_topten = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_topten, x= 'Product_ID', y= 'Orders')
```

## CONCLUSION

### Married Women in the age range of 26-35 who reside in the states of Uttar Pradesh, Maharashtra, and Karnataka, and are employed in the Information Technology (IT) sector, healthcare, and aviation industries, exhibit a higher propensity to engage in purchasing activities within the Food, Clothing, and Electronics categories.
